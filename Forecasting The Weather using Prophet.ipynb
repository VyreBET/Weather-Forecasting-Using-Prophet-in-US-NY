```python
import pandas as pd 
import numpy as np 

weather = pd.read_csv(r"D:\Projects\Projects\weather.csv", index_col="DATE")
```

```python
weather
```

```python
null_pct = weather.apply(pd.isnull).sum()/weather.shape[0]
null_pct 
```

```python
valid_columns = weather.columns[null_pct < .05]
valid_columns
```

```python
weather = weather[valid_columns].copy()
weather.columns = weather.columns.str.lower() ## THis is use to make the Columns or Horizontal Header to Small 
weather.index = pd.to_datetime(weather.index) # To store as Datetime not String!!!
weather 
```

```python
weather["station"].unique()
```

```python
## Reshaping the DataFrame 
lga = weather[weather["station"] == "USW00014732"].copy()
lga  
```

```python
weather = weather[weather["station"] == 'USW00094789'].copy()
weather
```

```python
weather = weather.merge(lga, left_index=True, right_index=True)
weather 
```

```python
weather["y"] = weather.shift(-1)["tmax_x"] 
weather[["tmax_x", "y"]]
```

```python
weather = weather.ffill()
weather 
```

```python
weather["ds"] = weather.index
weather
```

```python
predictors = weather.columns[~weather.columns.isin(["ds", "y", "station_x", "name_x", "station_y", "name_x"])]
predictors 
```

```python
train = weather[:'2021-12-31']
test = weather['2021-12-31':]
```

```python
from prophet import Prophet

# Only use numeric predictors for Prophet
numeric_predictors = [p for p in predictors if pd.api.types.is_numeric_dtype(train[p])]

def fit_prophet(train):
    m = Prophet()
    for p in numeric_predictors:
        m.add_regressor(p)
    m.fit(train)
    return m

m = fit_prophet(train)
```

```python
predictions = m.predict(test)
predictions
```

```python
from prophet.plot import plot_plotly, plot_components_plotly, plot_cross_validation_metric

plot_components_plotly(m, predictions)
```
